5.3 填充缺失值

```python
v = [10,None,30,40]
k = ['1号','2号','3号','4号']
sr1 = pd.Series(v, index=k)
print(sr1)
```

```python

print(sr1.fillna(0))
import numpy as np
print(sr1.fillna(np.mean(sr1)))
print(sr1.ffill())
print(sr1.bfill())
```

```python

v = [[None,1],[64,None],[72,3],[82,4]]
i = ['1号','2号','3号','4号']
c = ['年龄','牌照']
df1 = pd.DataFrame(v, index=i, columns=c)
print(df1)
```

```python
print(df1.bfill())
print(df1.ffill())
print(df1.fillna(2))
```
