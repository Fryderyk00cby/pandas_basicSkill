1.4二维对象的属性

```python
v= [[53,'男'],[64,'女'],[25,'女'],[44,'男'],[35,'女']]
i = ['1号','2号','3号','4号','5号']
c = ['年龄','性别']
df = pd.DataFrame(v,index=i,columns=c)
print(df)
```

```python
#values属性
print(df.values) #返回一个numpy数组（无论创建时是列表、数组还是张量）
#index属性
print(df.index) #返回一个Index对象，Index对象是不可变的，可以看成是一个特殊的数组
#columns属性
print(df.columns) #返回一个Index对象，Index对象是不可变的，可以看成是一个特殊的数组
```

```python
#提取年龄列
arr = df.values
arr = arr[:,0].astype(int)
print(arr)
```
