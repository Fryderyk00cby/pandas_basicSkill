3.3 对象的重塑

```python
#将sr并入df，也可以从df切分出sr
i = ['1号','2号','3号','4号','5号']
age = [53,64,25,44,35]
gender = ['男','女','女','男','女']
rank = [1,2,3,4,5]
sr1 = pd.Series(age,index=i)
sr2 = pd.Series(gender,index=i)
sr3 = pd.Series(rank,index=i)
print(sr1,'\n',sr2,'\n',sr3)
```

```python
#字典方式创建df
df = pd.DataFrame({'年龄':sr1,'性别':sr2})
print(df)
#用标签插入一个sr
df['排名'] = sr3
print(df)
```

```python
#从df中切分出一个sr
sr4 = df['年龄']
print(sr4)
```
