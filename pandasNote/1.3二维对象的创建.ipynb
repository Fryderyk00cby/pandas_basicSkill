1.3 二维对象的创建

```python
#创建一个二维对象存储病人的年龄和性别

#字典创建法
age = [53,64,25,44,35]
gender = ['男','女','女','男','女']
i = ['1号','2号','3号','4号','5号']
sr1 = pd.Series(age,index=i)
sr2 = pd.Series(gender,index=i)
df = pd.DataFrame({'年龄':sr1,'性别':sr2})
print(df)
#如果index不一致 会取交集
```

```python
#列表创建法
v= [[53,'男'],[64,'女'],[25,'女'],[44,'男'],[35,'女']]
i = ['1号','2号','3号','4号','5号']
c = ['年龄','性别']
df = pd.DataFrame(v,index=i,columns=c)
print(df)
```
