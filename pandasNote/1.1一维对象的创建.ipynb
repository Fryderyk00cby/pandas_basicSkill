```python
import pandas as pd
import numpy as np
import torch 
```

1.1 一维对象的创建

```python
#字典创建法
dict_v = {'a':0,'b':1,'c':2,'d':3,'e':4}
s = pd.Series(dict_v)
print(s)
dict_u = {'a':0,'b':0.25,'c':0.5,'d':0.75,'e':1}
t = pd.Series(dict_u)
print(t)
```

```python
#列表/张量/数组创建法
arr_v = [0,1,2,3,4]
arr_k = ['a','b','c','d','e']
s = pd.Series(arr_v,index=arr_k)
print(s)
arr_u = np.array([0,0.25,0.5,0.75,1])
t = pd.Series(arr_u,index=arr_k)
print(t)
arr_w = torch.tensor([0,0.25,0.5,0.75,1])
w = pd.Series(arr_w,index=arr_k)
print(w)
```
