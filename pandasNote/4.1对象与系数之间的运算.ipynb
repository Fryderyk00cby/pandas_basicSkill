4.1 对象与系数之间的运算

```python
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s)
```

```python
s = s + 10
print(s)
s = s **2
print(s)
```

```python
v = [[53,'男'],[64,'女'],[25,'女'],[44,'男'],[35,'女']]
df = pd.DataFrame(v, index=['a','b','c','d','e'], columns=['年龄','性别'])
print(df)
print(df['年龄'].values)
```

```python
df['年龄'] = df['年龄'] + 10
print(df)
```

```python
df['年龄'] = df['年龄'] ** 2
print(df)
```
