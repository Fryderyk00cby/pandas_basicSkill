```python
import pandas as pd
```

2.1 一维对象的索引

```python
#pandas中索引有两种 loc和iloc
#loc是基于标签的索引，iloc是基于位置的索引
arr_v = [0,1,2,3,4]
arr_k = ['a','b','c','d','e']
s = pd.Series(arr_v,index=arr_k)
```

```python

#访问元素
print(s.loc['c']) #基于标签的索引
print(s.iloc[2]) #基于位置的索引
#花式索引
print(s.loc[['a','c','e']]) #基于标签的索引
print(s.iloc[[0,2,4]]) #基于位置的索引
#修改元素
s.loc['a'] = 10 #基于标签的索引
s.iloc[1] = 20 #基于位置的索引
s.iloc[[2,3,4]] = [30,40,50] #花式索引修改
print(s)
```

```python
#访问切片
print(s.loc['b':'d']) #基于标签的切片，包含结束标签
print(s.iloc[1:4]) #基于位置的切片，不包含结束位置
#切边仅是视图 对象赋值仅是绑定（与numpy相同）
s1 = s.loc['b':'d'] #基于标签的切片，包含结束标签
s1.loc['b'] = 200 #修改s1中的元素
print(s) #s中的元素也被修改了
s2 = s.iloc[1:4] #基于位置的切片，不包含结束位置
s2.iloc[0] = 300 #修改s2中的元素
print(s) #s中的元素也被修改了
```
