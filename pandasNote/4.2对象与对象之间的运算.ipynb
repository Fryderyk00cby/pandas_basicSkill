```python
import pandas as pd
```

```python
v1 = [10, 20, 30, 40, 50]
k1 = ['a', 'b', 'c', 'd', 'e']
s1 = pd.Series(v1, index=k1)
print(s1)
```

```python
v2 = [1, 2, ]
k2 = ['a', 'b']
s2 = pd.Series(v2, index=k2)
print(s2)
```

```python
print(s1 + s2)
print(s1 - s2)
print(s1 * s2)
print(s1 / s2)
```

```python
#二维对象之间的运算
v1 = [[53,'男'],[64,'女'],[25,'女'],[44,'男']]
v2 = [1,2,3,6]
i1 = ['1号','2号','3号','4号']
c1 = ['年龄','性别']
i2 = ['1号','2号','3号','6号']
c2 = ['牌照']
df1 = pd.DataFrame(v1, index=i1, columns=c1)
df2 = pd.DataFrame(v2, index=i2, columns=c2)
print(df1)
print(df2)
```

```python
df1['加法'] = df1['年龄'] + df2['牌照']
print(df1)
df1['减法'] = df1['年龄'] - df2['牌照']
print(df1)
df1['乘法'] = df1['年龄'] * df2['牌照']
print(df1)
df1['除法'] = df1['年龄'] / df2['牌照']
print(df1)
```

```python
#也可以用numpy数学函数
import numpy as np
df1['年龄2'] = np.sqrt(df1['年龄'])
print(df1)
print(df1['年龄2'].values)
```

```python
#bool型数组
bool_array = df1['年龄'] > 30
print(bool_array)
print(np.sum(bool_array))
```

```python
#bool作为索引查找
print(df1[bool_array ==False])
```
