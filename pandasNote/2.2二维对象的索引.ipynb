2.2 二维对象的索引

```python
age = [53,64,25,44,35]
gender = ['男','女','女','男','女']
i = ['1号','2号','3号','4号','5号']
sr1 = pd.Series(age,index=i)
sr2 = pd.Series(gender,index=i)
df = pd.DataFrame({'年龄':sr1,'性别':sr2})
print(df)
```

```python
#访问元素
print(df.loc['1号', '年龄']) #基于标签访问
print(df.loc['1号', '性别']) #基于标签访问
print(df.iloc[0, 0]) #基于位置访问
print(df.iloc[0, 1]) #基于位置访问

print(df.loc[['1号','3号'], ['性别', '年龄']]) #基于标签访问多个元素
print(df.iloc[[0,2], [1,0]]) #基于位置访问多个
```

```python
#切片
print(df.loc['1号':'3号', '年龄':'性别']) #基于标签的切片，包含结束标签
print(df.iloc[3:0:-1, 0:2]) #基于位置的切片，不包含结束位置
```

```python
#提取一个特征的数据
print(df['年龄']) #提取一个特征的数据 相当于df.loc[:, '年龄'] 即为一个Series对象
print(df['年龄']['1号']) #提取一个特征的数据中的一个元素
print(df['年龄'].values) #提取一个特征的数据中的所有元素，返回一个numpy数组
```
