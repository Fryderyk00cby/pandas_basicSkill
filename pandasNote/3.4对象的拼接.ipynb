```python
import pandas as pd
```

3.4 对象的拼接

```python
v1 = [10,20,30,40]
i1 = ['1号','2号','3号','4号']
sr1 = pd.Series(v1, index=i1)
print(sr1)
v2 = [50,60,70,80]
i2 = ['5号','6号','7号','8号']
sr2 = pd.Series(v2, index=i2)
print(sr2)
```

```python
#对象的拼接
sr = pd.concat([sr1, sr2])
print(sr)
```

```python
#df与sr的拼接 即给df添加一列或者一行 用索引即可
df = pd.DataFrame({'年龄':sr1})
print(df)
df['score'] = [20,30,40,50]
print(df)
df.loc['5号'] = [20,60]
print(df)
```

```python
#二维对象的拼接
v1 = [[10,'女'],[20,'男'],[30,'女'],[40,'男']]
v2 = [[1,'是'],[2,'否'],[3,'是'],[4,'否']]
v3 = [[50,'男',5,'是'],[60,'女',6,'否']]
i1 = ['1号','2号','3号','4号']
i2 = ['1号','2号','3号','4号']
i3 = ['5号','6号']
c1 = ['年龄','性别']
c2 = ['排名','ikun']
c3 = ['年龄','性别','排名','ikun']
df1 = pd.DataFrame(v1, index=i1, columns=c1)
print(df1)
df2 = pd.DataFrame(v2, index=i2, columns=c2)
print(df2)
df3 = pd.DataFrame(v3, index=i3, columns=c3)
print(df3)
```

```python
#合并列对象
df4 = pd.concat([df1, df2], axis=1)
print(df4)
```

```python
#合并行对象
df5 = pd.concat([df4, df3], axis=0)
print(df5)
```
