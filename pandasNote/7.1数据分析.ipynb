7.1 数据分析

```python
df = pd.read_csv('test_table.csv', index_col=0)
df.head() #显示前五行数据
```

```python
#numpy聚合函数均可以在pandas中使用
print(df.mean()) #求平均值
print(df.sum()) #求和
print(df.max()) #求最大值
print(df.min()) #求最小值
print(df.std()) #求标准差
print(df.describe()) #一次性显示所有统计指标
print(df['年龄'].mean()) #求年龄列的平均值
print(df['年龄'].sum()) #求年龄列的和
```

```python
#数据透视
df_titanic = pd.read_csv('train.csv')
print(df_titanic.head())
```

```python
#一个特征
print(df_titanic.pivot_table('Survived', index='Sex'))#默认aggfunc为mean
```

```python
#两个特征
print(df_titanic.pivot_table('Survived', index='Sex', columns='Pclass'))
```

```python
#多特征数据透视表
age = pd.cut(df_titanic['Age'], [0, 25, 80])
print(age)
print(df_titanic.pivot_table('Survived', index=['Sex', age], columns='Pclass',observed=False))
```

```python
#四维数据透视表
fare = pd.qcut(df_titanic['Fare'], 2)
print(fare)
print(df_titanic.pivot_table('Survived', index=['Sex', age], columns=[fare, 'Pclass'], observed=False)) 
```
